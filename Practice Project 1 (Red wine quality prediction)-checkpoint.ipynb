# Preprocessing the data

**Dropping columns which are having very very low correlation with our quality or target and having outliers too and skewed also**

```python
df=df.drop(['residual sugar','chlorides','free sulfur dioxide', ],axis=1)
```

```python
df
```

```python
df.shape
```

```python
x=df.iloc[:,0:-2]
y=df.iloc[:,-1]
```

```python
x.shape
```

```python
y.shape
```

**removing outliers fromdataset**

```python
# by Zscore method
# importing required libraries
from scipy.stats import zscore
```

```python
z=np.abs(zscore(x))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
x=x[(z<3).all(axis=1)]
x.shape
```

```python
y=y[(z<3).all(axis=1)]
y.shape
```

```python
# calculating percentage data loss
loss=(1599-1510)/1599*100
loss
```

**the data loss is 5.56% is may be fall under acceptable range**

**Removing Skewness** using PowerTransform

```python
# Importing library to use PoerTransform
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
x
```

**Scaling the data**

```python
# importing libraries
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

**Bancing data by oversampling**

```python
y.value_counts()
```

**As we can see that the data is imbalance we can balance it by imblearn over_sampling**

```python
# Importing Libraries for it 
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE()
x,y=sm.fit_resample(x,y)
```

```python
y.value_counts()
```

**we have converted the multi categorical output in only two categories as mentioned above so Binary Classification model can be used**

# Modelling the dataset

```python
# importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
lr=LogisticRegression()
```

**To find best random state**

```python
maxaccu=0
maxrs=0
for i in range(1,1000):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.20, random_state=i)
    lr.fit(xtrain,ytrain)
    predlr=lr.predict(xtest)
    acc=accuracy_score(ytest,predlr)
    print('testing accuracy', acc, 'randon state',i)
    
    if acc>maxaccu:
        maxaccu=acc
        maxrs=i
        print('max accuracy', maxaccu , 'max_random_state', i)
    
```

```python
print('best accuracy is ',maxaccu,'best randon state is ',maxrs)
```

**Creating the train test split according to our best random state**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=.20, random_state=301)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

**using now Decision tree classifier**

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

**using Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=200)
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

**using SVC**

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

**using KNN Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
predknn=knn.predict(x_test)
print(accuracy_score(y_test,predknn))
print(confusion_matrix(y_test,predknn))
print(classification_report(y_test,predknn))
```

**using GradientBoostingClassifier**

```python
from sklearn.ensemble import GradientBoostingClassifier
xgb=KNeighborsClassifier()
xgb.fit(x_train,y_train)
predxgb=xgb.predict(x_test)
print(accuracy_score(y_test,predxgb))
print(confusion_matrix(y_test,predxgb))
print(classification_report(y_test,predxgb))
```

**using Adaboostclassifier**

```python
from sklearn.ensemble import AdaBoostClassifier
ada=KNeighborsClassifier()
ada.fit(x_train,y_train)
predada=ada.predict(x_test)
print(accuracy_score(y_test,predada))
print(confusion_matrix(y_test,predada))
print(classification_report(y_test,predada))
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score

# validation accuracy
model=[lr,dtc,rf,svc,knn,xgb,ada]

for i in model:
    for j in range(2,10):
        print('taking cv =', j)
        cr_score=cross_val_score(i, x, y, cv=j)
        print('Cross validation score for ' , i ,' is', cr_score)
        print('Mean of Cross Val. score is ', cr_score.mean())
    
```

**It is observed that by taking cv = 7 and with Random Forest Classifier the model is performing good** 

# Hyper Parameter tuning

```python
from sklearn.model_selection import GridSearchCV

#Creating parameter list to be passed in GridSearchCV

parameters={'max_features':['auto','sqrt','log2'] , 
            'max_depth':[4,5,6,7,8],
            'criterion':['gini','entropy'], 
            'n_estimators': [100,200,300,400,500]}
```

```python
GCV=GridSearchCV(estimator=RandomForestClassifier(),
                param_grid=parameters,
                cv=7,
                scoring='accuracy')
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python
GCV.best_estimator_
```

```python
final_model=RandomForestClassifier(criterion='gini',max_depth= 8, max_features='log2', n_estimators= 400)
final_model.fit(x_train,y_train)
predfinal=final_model.predict(x_test)
print('Final Accuracy is with Random Forest Classifier = ',accuracy_score(y_test,predfinal))
```
