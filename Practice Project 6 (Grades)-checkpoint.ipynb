Problem Statement:
The dataset contains grades scored by students throughout their university tenure in various courses and their CGPA calculated based on their grades

```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Loding dataset to jupyter
df=pd.read_csv("https://raw.githubusercontent.com/dsrscientist/dataset4/main/Grades.csv")
df
```

The columns succeeding the seat no. are the subjects in the entire course in university with last column as the CGPA offered to student.

**Observation**
1. As we can see that all the columns except the CGPA column all are the object type which is right.
2. Also we can see that there cannot be any void present in the data as data types are right.

```python
#Checking the Null values
df.isnull().sum()
```

There are so many null values are present.
for a categorical data the null values can be filled with the mode of that column

```python
# Using for loop for filling the null values for the columns
for i in df.columns:
    df[i]=df[i].fillna(df[i].mode()[0])
df    
```

```python
# Cheking the values filled or not
df.isnull().sum()
```

```python
# Checking for the unique values of each columns to deal with it
for i in df.columns:
    print(df[i].unique())
```

**Observations related with the database containing the special characters**
1. The symbol “WU” indicates that an enrolled student did not officially withdraw from the course and also failed to complete course requirements
2. The symbol “W” indicates that the student was permitted to drop the course
3. The symbol “I” indicates that a portion of required coursework has not been completed
**For all those above conditions we can assign F grade as per understanding of the course result calculations**

```python
# Replacing WU (Withdrawal Unauthorized) as F because it that student is on roll course is not completed
for i in df.columns:
    df[i] = df[i].astype(str).str.replace("WU",'F')
```

```python
# Replacing W (Withdrawal) as F because it that student is on roll course is not completed
for i in df.columns:
    df[i] = df[i].astype(str).str.replace("W",'F')
```

```python
# Replacing I (Incomplete) as F because it that student is on roll course is not completed
for i in df.columns:
    df[i] = df[i].astype(str).str.replace("I",'F')
```

```python
#Again checking for the unique values to ensure that the values containing special characters are gone or not
for i in df.columns:
    print(df[i].unique())
```

# Pre-Processing

```python
# Dividing the data in to inputs and outputs
x=df.drop(["Seat No.","CGPA"],axis=1)
x
```

```python
y=df["CGPA"]
y
```

```python
# Dealing with the outliers
from scipy.stats import zscore
z=np.abs(zscore(x))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
x=x[(z<3).all(axis=1)]
x.shape
```

```python
y=y[(z<3).all(axis=1)]
y.shape
```

```python
# calculating percentage data loss
loss=(571-540)/571*100
loss
```

**As the data loss is less than 10% we can remove outliers**

```python
# Checking for skewness
x.skew().sort_values(ascending=False)
```

**for multiple columns skewness is observed we can remove skewness by power transform**

```python
# Removing skewness
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
x
```

```python
# checking again whether skewness is removed or not
sk=pd.DataFrame(x)
sk.skew()
```

**As we can see that the skewness is removed from the data**

```python
#scaling the data
# importing libraries
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
#importing the VIF in jupyter
from statsmodels.stats.outliers_influence import variance_inflation_factor
new_df=pd.DataFrame(x)
def vif_scores(df):
    VIF_Scores = pd.DataFrame()
    VIF_Scores["Independent Features"] = new_df.columns
    VIF_Scores["VIF Scores"] = [variance_inflation_factor(new_df.values,i) for i in range(new_df.shape[1])]
    return VIF_Scores
df1 = new_df
vif_scores(df1)
```

**As it is clear from data VIF score the scores are less than 5 we do not treat for the multcolinearity**

# Modelling the data

**The target column is of continuous type hence we can apply the linear regression**

```python
# importing Linear Regression and metrics for it
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

Finding out best random state with linear regression model

```python
maxr2=0
maxrs=0
lr=LinearRegression()
for i in range(0,200):
    print("Random State is ", i)
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=i)
    lr.fit(xtrain,ytrain)
    predtrain=lr.predict(xtrain)
    predtest=lr.predict(xtest)
    print('training accuracy is ', r2_score(ytrain,predtrain))
    score=r2_score(ytest,predtest)
    print('testing accuracy is ', score)
    print("\n")
    
    if score>maxr2:
        maxr2=score
        maxrs=i
        
        
print("maximum r2 score is ", maxr2 , "at random state ", maxrs)
```

**the maximum r2 score is  0.9634696356865128 at random state  166**

Spliting the model into train test split according to above result

```python
#Train Test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=166)
```

```python
#running that same model again for other metrics too
lr.fit(x_train,y_train)
predtrainlr=lr.predict(x_train)
predtestlr=lr.predict(x_test)
print('training accuracy is ', r2_score(y_train,predtrainlr))
print('testing accuracy is ', r2_score(y_test,predtestlr))
print("Mean Squared Error is ",mean_squared_error(y_test,predtestlr))
print("Mean Absolute Error is ",mean_absolute_error(y_test,predtestlr))
print("Root Mean Squared Error is ",np.sqrt(mean_squared_error(y_test,predtestlr)))
```

```python
from sklearn.model_selection import cross_val_score
#Cross validation of the lr model
for j in range(2,10):
    cv_score=cross_val_score(lr, x, y, cv=j)
    print('At fold ', j , "Mean CV score is ", cv_score.mean())
    print('training accuracy is ', r2_score(y_train,predtrainlr))
    print('testing accuracy is ', r2_score(y_test,predtestlr))
```

**we can see that model is giving same cross val scores for folds more than 3**
Hence selecting the 5 folds

# Regularization

```python
from sklearn.linear_model import Lasso,Ridge
from sklearn.model_selection import GridSearchCV
```

```python
# Doing Lasso Regularization
#HyperParameter Tunning to find out the best parameters
parameters={'alpha':[.00001,0.0001,0.001,0.01,0.1,1,10,100],
            'random_state':list(range(0,10)),
            'max_iter':[10,100,1000,10000],
            'tol': [0.001,0.0001,0.00001,0.000001]}

ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
ls=Lasso(alpha=0.01, max_iter=100, random_state=0, tol=1e-05)
ls.fit(x_train,y_train)
predtrainls=ls.predict(x_train)
predtestls=ls.predict(x_test)
print('training accuracy is ', r2_score(y_train,predtrainls))
print('testing accuracy is ', r2_score(y_test,predtestls))
print("Mean Squared Error is ",mean_squared_error(y_test,predtestls))
print("Mean Absolute Error is ",mean_absolute_error(y_test,predtestls))
print("Root Mean Squared Error is ",np.sqrt(mean_squared_error(y_test,predtestls)))
```

```python
#Cross validation of the ls model
for j in range(2,10):
    cv_score=cross_val_score(ls, x, y, cv=j)
    print('At fold ', j , "Mean CV score is ", cv_score.mean())
```

```python
# Doing Ridge Regularization
#HyperParameter Tunning to find out the best parameters
parameters={'alpha':[.00001,0.0001,0.001,0.01,0.1,1,10,100],
            'random_state':list(range(0,10)),
            'max_iter':[10,100,1000,10000],
            'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga', 'lbfgs']}

rg=Ridge()
clf=GridSearchCV(rg,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
rg=Ridge(alpha=100, max_iter=10, random_state=5, solver='sag')
rg.fit(x_train,y_train)
predtrainrg=rg.predict(x_train)
predtestrg=rg.predict(x_test)
print('training accuracy is ', r2_score(y_train,predtrainrg))
print('testing accuracy is ', r2_score(y_test,predtestrg))
print("Mean Squared Error is ",mean_squared_error(y_test,predtestrg))
print("Mean Absolute Error is ",mean_absolute_error(y_test,predtestrg))
print("Root Mean Squared Error is ",np.sqrt(mean_squared_error(y_test,predtestrg)))
```

```python
#Cross validation of the ls model
for j in range(2,10):
    cv_score=cross_val_score(rg, x, y, cv=j)
    print('At fold ', j , "Mean CV score is ", cv_score.mean())
```
